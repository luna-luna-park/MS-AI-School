```python
n = 10
for i in range(1, n + 1):
    # 첫 번째 내부 for 루프: 각 행의 왼쪽 공백 출력
    for j in range(n - i):
        print(' ', end='')
    # 두 번째 내부 for 루프: 각 행에 대한 별 출력
    for j in range(2 * i - 1):
        print('*', end='')
    # 한 행이 끝나면 줄 바꿈
    print()
```

#### 끝말잇기 게임

```python
# 끝말잇기 게임

다리미 -> 미소 -> 소방소 -> 강아지 -> 땡 게임종료 
```

```python
print('끝말잇기 게임 시작단어: 다리미')
word = '다리미'

# print(word[0], word[-1])
while True:
    print('while 문 안')
    answer = input('단어를 입력하세요: ')
    print('input 입력 받은 후')
    print('단어를 입력하세요: ', answer)
    # 끝말잇기 단어 비교하기
    if word[-1] == answer[0]: #끝말잇기 조건을 만족했다면 
        print(word[-1],answer[0] )
        word = answer #다음 단계로 넘어가기     
        print(word)  
    else:
        print("땡")
        # while문 종료
        break 

print('게임종료')
```

```python
addr = {
    '박지민': '010-111-1111',
    '전정국': '010-222-2222'
}

while True:
    menu = int(input('메뉴선택(1: 추가, 2: 검색, 0: 종료)? '))

    if menu == 1:
        name = input('이름입력: ')
        phone = input('전화번호 입력: ')
        addr[name] = phone
        print(f'{name}의 연락처 {phone} 추가')
    elif menu==2:
        print('검색가능 친구 목록: ', list(addr.keys()))
        name = input('이름입력: ')
        if addr.get(name):
            print(f'{name}의 전화번호: ', addr[name])
        else:
            print('{name} 그런 사람은 없습니다.')
    elif menu == 0:
        print('프로그램 종료')
        break
    else:
        print('메뉴선택 오류')
        continue
```
